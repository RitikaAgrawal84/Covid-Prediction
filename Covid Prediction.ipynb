```python
import pandas as pd #importing pandas to work with dataframe
import seaborn as sns #To make plots
import matplotlib.pyplot as plt # To display plots 
%matplotlib inline 
sns.set(color_codes=True) #To set the default color palette for plots 
```

```python
covid = data.drop(columns=["ID","sex","age","pcr_date","exam_bh_date"])
```

```python
from sklearn.preprocessing import LabelEncoder

# Creating an instance of LabelEncoder
le = LabelEncoder()

# Fit and transform the categorical data
# encoding the 'diagnosis' column from categorical to numerical
covid['pcr'] = le.fit_transform(covid['pcr'])
```

```python
fig,ax=plt.subplots(figsize=(18,18))
# creating a heatmap of the correlation matrix
dataplot = sns.heatmap(covid.corr(),cmap='coolwarm',annot=True) 
```

```python
# creating feature matrix X by dropping the 'diagnosis' column
X = covid.drop(columns="pcr",axis=1) 

# Creating target vector y containing the 'diagnosis' column
y = covid["pcr"] 
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB

# Split data into training and test sets (80% training, 20% testing) using stratified sampling for the PCR class
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Standardize the feature variables
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize the models
rfc_model = RandomForestClassifier()
logistic_model = LogisticRegression()
svc_model = SVC(kernel="poly",degree=1)
naive_bayes_model = GaussianNB()

# Train the models
rfc_model.fit(X_train_scaled, y_train)
logistic_model.fit(X_train_scaled, y_train)
svc_model.fit(X_train_scaled, y_train)
naive_bayes_model.fit(X_train_scaled, y_train)

# Make predictions on the test set
rfc_predictions = rfc_model.predict(X_test_scaled)
logistic_predictions = logistic_model.predict(X_test_scaled)
svc_predictions = svc_model.predict(X_test_scaled)
naive_bayes_predictions = naive_bayes_model.predict(X_test_scaled)

# Calculate and print accuracy scores for each model
rfc_accuracy = accuracy_score(y_test, rfc_predictions)
logistic_accuracy = accuracy_score(y_test, logistic_predictions)
svc_accuracy = accuracy_score(y_test, svc_predictions)
naive_bayes_accuracy = accuracy_score(y_test, naive_bayes_predictions)

print(f"Random Forest Classifier Accuracy: {rfc_accuracy:.2%}")
print(f"Logistic Regression Accuracy: {logistic_accuracy:.2%}")
print(f"Support Vector Classifier Accuracy: {svc_accuracy:.2%}")
print(f"Naive Bayes Classifier Accuracy: {naive_bayes_accuracy:.2%}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Calculate confusion matrix for each model
rfc_confusion_matrix = confusion_matrix(y_test, rfc_predictions)
logistic_confusion_matrix = confusion_matrix(y_test, logistic_predictions)
svc_confusion_matrix = confusion_matrix(y_test, svc_predictions)
naive_bayes_confusion_matrix = confusion_matrix(y_test, naive_bayes_predictions)

# Define a function to plot a confusion matrix with smaller labels in a subplot
def plot_confusion_matrix(ax, conf_matrix, model_name):
    # Plot the confusion matrix using seaborn heatmap
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=['Neg', 'Pos'],  # Shortened labels
                yticklabels=['Neg', 'Pos'],  # Shortened labels
                annot_kws={'size': 12}, ax=ax)  # Smaller font size for annotations
    
    # Set x and y axis labels and title with smaller font size
    ax.set_xlabel('Predicted', fontsize=12)
    ax.set_ylabel('Actual', fontsize=12)
    ax.set_title(f'Confusion Matrix for {model_name}', fontsize=14)

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(8, 8))

# Plot confusion matrices for each model in the grid
plot_confusion_matrix(axes[0, 0], rfc_confusion_matrix, 'Random Forest Classifier')
plot_confusion_matrix(axes[0, 1], logistic_confusion_matrix, 'Logistic Regression')
plot_confusion_matrix(axes[1, 0], svc_confusion_matrix, 'Support Vector Classifier')
plot_confusion_matrix(axes[1, 1], naive_bayes_confusion_matrix, 'Naive Bayes Classifier')

# Adjust layout and spacing between plots for better visualization
plt.subplots_adjust(wspace=1, hspace=0.7)

# Show the plots
plt.show()
```

```python
from sklearn.metrics import classification_report

# Define a function to calculate and display sensitivity (recall) and specificity for a given model
def display_metrics(model_name, y_true, y_pred):
    # Calculate classification report
    report = classification_report(y_true, y_pred, target_names=['Neg', 'Pos'], output_dict=True)
    
    # Extract sensitivity and specificity
    sensitivity = report['Pos']['recall']
    specificity = report['Neg']['recall']
    
    # Display metrics
    print(f"Model: {model_name}")
    print(f"Sensitivity (Recall): {sensitivity:.2f}")
    print(f"Specificity: {specificity:.2f}")
    print()

# Display metrics for each model
display_metrics('Random Forest Classifier', y_test, rfc_predictions)
display_metrics('Logistic Regression', y_test, logistic_predictions)
display_metrics('Support Vector Classifier', y_test, svc_predictions)
display_metrics('Naive Bayes Classifier', y_test, naive_bayes_predictions)
```

```python
def predict_new_values(new_values):
    # Convert new_values list to a 2D array as expected by the scaler
    new_values_reshaped = [new_values]
    
    # Scale the new values using the same scaler used during training
    new_values_scaled = scaler.transform(new_values_reshaped)
    
    # Make predictions using each model
    rfc_prediction = rfc_model.predict(new_values_scaled)
    logistic_prediction = logistic_model.predict(new_values_scaled)
    svc_prediction = svc_model.predict(new_values_scaled)
    naive_bayes_prediction = naive_bayes_model.predict(new_values_scaled)
    
    # Return predictions from each model
    return {
        'Random Forest Classifier': 'Positive' if rfc_prediction[0] == 1 else 'Negative',
        'Logistic Regression': 'Positive' if logistic_prediction[0] == 1 else 'Negative',
        'Support Vector Classifier': 'Positive' if svc_prediction[0] == 1 else 'Negative',
        'Naive Bayes Classifier': 'Positive' if naive_bayes_prediction[0] == 1 else 'Negative'
    }

# Example usage of the function
new_input = [1550,70.3,10.9,40.1,0.3,1,130,1600,8500,20,30,6500,190.2,89.8,34.1,7.5,1.4,25.1,1000,11.5]

predictions = predict_new_values(new_input)

# Print the predictions made by each model
for model_name, prediction in predictions.items():
    print(f'{model_name} Prediction: {prediction}')

```
